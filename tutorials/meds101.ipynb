```python
import fitsio

from ngmix.medsreaders import NGMixMEDS

import proplot as pplt

import numpy as np
```

```python
m = NGMixMEDS("DES0954+0252_r4918p01_i_meds-DECADE.fits.fz")
```

```python
i = 11
obslist = m.get_obslist(i)
```

```python
fig, axs = pplt.subplots(nrows=3, ncols=3, share=0)

for i in range(3):
    obs = obslist[i]
    axs[0, i].imshow(np.arcsinh(obs.image * np.sqrt(obs.weight)), cmap='rocket')
    axs[1, i].imshow(obs.bmask, cmap='rocket')
    axs[2, i].imshow(np.arcsinh(obs.psf.image[10:-10, 10:-10]), cmap='rocket')

axs[0, 0].set_title("coadd")
axs[0, 1].set_title("single-epoch 1")
axs[0, 2].set_title("single-epoch 2")

axs[0, 0].set_ylabel("image")
axs[1, 0].set_ylabel("bmask")
axs[2, 0].set_ylabel("PSF")

axs.grid(False)
```
